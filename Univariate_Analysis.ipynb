```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path='/content/apy_1.csv'
df=pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
df = df.dropna()
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
df.hist(figsize=(10, 8), bins=20, color='orange', edgecolor='black')
plt.suptitle('Histograms of crop production')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x='Season', y='State_Name', data=df)
plt.title('Box Plots of State name by Season')
plt.xlabel('Season')
plt.ylabel('State_Name')
plt.show()
```

```python
df.plot(kind='density', subplots=True, layout=(2,2), sharex=False, figsize=(10,8))
plt.suptitle('Density Plots of crop production')
plt.show()
```
